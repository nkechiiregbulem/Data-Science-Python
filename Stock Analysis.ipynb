# Technical Indicators

```python
from alpha_vantage.techindicators import TechIndicators
import matplotlib.pyplot as plt

ti = TechIndicators(key='7TOTGF7E6UZ6ESJD', output_format='pandas')
data, meta_data = ti.get_rsi(symbol='CME', interval='60min', time_period=60)
data.plot()
plt.title('RSI indicator for CME stock (60 min)')
plt.show()
```

```python
from alpha_vantage.techindicators import TechIndicators
import matplotlib.pyplot as plt

ti = TechIndicators(key='7TOTGF7E6UZ6ESJD', output_format='pandas')
data, meta_data = ti.get_macd(symbol='CME', interval='60min', series_type='high')
data.plot()
plt.title('MACD indicator for CME stock (60 min)')
plt.show()
```

# Sector Performance

```python
from alpha_vantage.sectorperformance import SectorPerformances
import matplotlib.pyplot as plt

sp = SectorPerformances(key='7TOTGF7E6UZ6ESJD', output_format='pandas')
data, meta_data = sp.get_sector()
data['Rank A: Real-Time Performance'].plot(kind='bar')
plt.title('Real Time Performance (%) per Sector')
plt.tight_layout()
plt.grid()
plt.show()
```

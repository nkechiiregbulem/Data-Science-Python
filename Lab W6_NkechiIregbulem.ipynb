<b><font color="blue", size = 4>a)</font></b> What were the 15 most recent technology IPOs on `NYSE`?

<span style="background-color: #FFFF00">Your answer here</span>

```python
exchange_columns = ['Symbol', 'Name', 'LastSale', 'MarketCap', 'IPOyear', 'Sector', 'Industry']
```

```python
url = "https://www.nasdaq.com/screening/companies-by-industry.aspx?exchange=NYSE&render=download"
try:
    NYSE = pd.read_csv(urlopen(url))
    NYSE = NYSE[exchange_columns]
    NYSE = NYSE.replace('n/a',np.NaN).dropna()
    display(NYSE.head(10))
except:
    print ("Error with the following url")
    print (url)
```

```python
tech_df = NYSE[NYSE.Sector == "Technology"].sort_values("IPOyear", ascending = False)[0:10]
tech_df
```
